```python
import pandas as pd
```

```python
df = pd.read_csv('data/data_final/pre_summ.csv')
```

```python
df
```

```python
df.drop(df.columns[[0]], axis = 1, inplace=True)
```

```python
df
```

```python
from nltk.tokenize import sent_tokenize
```

```python
more = 0
less = 0
for i in range(len(df)):
    text = df['reviewText'][i]
    tokenized_text = sent_tokenize(text)
    if(len(tokenized_text) >= 15):
        more = more + 1
    else:
        less = less + 1

print(more, less)
```

```python
df
```

```python
df2 = pd.DataFrame(columns = ['reviewText', 'summary', 'review_clean', 'sentiment'])
```

```python
for i in range(len(df)):
    text = df['reviewText'][i]
    tokenized_text = sent_tokenize(text)
    if(len(tokenized_text) < 15):
        df.drop([i], inplace=True)

print(df.shape)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('index', axis='columns', inplace=True)
```
